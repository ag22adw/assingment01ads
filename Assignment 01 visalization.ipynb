```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('goldsilver.csv')

# Define function to create line plot
def create_line_plot(x_data, y_data, title, x_label, y_label):
    plt.plot(x_data, y_data)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

# Create line plot for gold prices
create_line_plot(df['Year'], df['Gold Price '], 'Gold Prices 1791-2021', 'Year', 'Price ($)')

# Create line plot for silver prices
create_line_plot(df['Year'], df['Silver Price '], 'Silver Prices 1791-2021', 'Year', 'Price ($)')

# Create line plot for both gold and silver prices
plt.plot(df['Year'], df['Gold Price '], label='Gold')
plt.plot(df['Year'], df['Silver Price '], label='Silver')
plt.title('Gold and Silver Prices 1791-2021')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')

purchase_by_age = df.groupby('Age')['Purchase'].sum()

purchase_by_age.plot(kind='bar')
plt.title('Total Purchase Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase Amount')
plt.show()
```
